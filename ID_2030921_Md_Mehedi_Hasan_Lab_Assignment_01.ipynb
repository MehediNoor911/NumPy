##### **Problem 02:**
Find the sum and mean of all the rows of the matrix.

```python
# Write your code here
for i in range(len(rand_matrix)):
  sum = 0
  for j in rand_matrix[i, :]:
    sum = sum + j
  mean = sum/len(rand_matrix[0])
  print(f"{i+1}th row's sum is {sum} and mean is {mean}")
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
# Write your code here
x = np.arange(-3, 3, 0.5)
y = np.arange(-3, 3, 0.5)
X,Y = np.meshgrid(x,y)
Z = 1-X-Y
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
ax.plot_surface(X, Y, Z)
plt.show()

```
